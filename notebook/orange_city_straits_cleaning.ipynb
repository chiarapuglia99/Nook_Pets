# Importo le librerie necessarie

```python
import pandas as pd
```

# Controllo la locazione del dataset per verificarne la correttezza

```python
df = pd.read_csv("../dataset/strade_orange_city.csv", encoding='latin-1')

# (Opzionale) Visualizza le informazioni iniziali per vedere quanti null ci sono
print("Info dataset originale:")
print(df.info())
```

# Eliminazione delle strade che presentano il valore del campo "name" uguale a null

Effettuo questa modifica poichè su qgis, è stata effettuata una categorizzazione delle strade e sono state eliminate alcune strade poichè non erano fondamentali ai fini della nostra analisi

```python
# 2. Eliminazione delle righe con valore 'name' nullo
df_cleaned = df.dropna(subset=['name'])
```

# Verifica ulteriore per assicurarsi che la pulizia sia stata effettuata correttamente

```python
print("\nInfo dataset pulito:")
print(df_cleaned.info())
```

# Salvataggio del dataset pulito in un nuovo file CSV

```python
output_filename = '../dataset/strade_orange_city_cleaned.csv'
df_cleaned.to_csv(output_filename, index=False)

print(f"\nFile salvato come: {output_filename}")
```

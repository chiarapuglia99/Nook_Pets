# Importo le librerie necessarie

```python
import pandas as pd
```

# Caricamento del dataset

```python
df = pd.read_csv("D:/Progetti/Nook_Pets/dataset/animali-rifugi.csv")
```

# Creazione della colonna "Health"
Creazione della colonna "Health" per indicare se un animale è stato castrato o sterilizzato in base al sesso

```python
df['Health'] = None
```

# Ricerca nella colonna "Sex" dei valori "Spayed"e "Neutered" e trasferimento nella colonna "Health"

```python
mask = df['Sex'].isin(['Spayed', 'Neutered'])


df.loc[mask, 'Health'] = df.loc[mask, 'Sex']
print(df)
```

# Sostituire in Sex i valori "Spayed" e "Neutered" con "Male" e Female in base alla colonna Health

```python
df.loc[df['Health'] == 'Spayed', 'Sex'] = 'Female'
df.loc[df['Health'] == 'Neutered', 'Sex'] = 'Male'

df[['Sex', 'Health']]
```

# Controllo delle colonne Sex e Health se ci sono valori Null/None oppure se ci sono ancora valori "Spayed" e "Neutered"

```python
print("\n1. CONTROLLO VALORI MANCANTI (Null/None)")

# Conta i nulli per colonna
null_counts = df[['Sex', 'Health']].isnull().sum()

if null_counts.sum() > 0:
    print(f"Attenzione: Trovati valori Null!\n{null_counts[null_counts > 0]}")

    # Mostra le righe specifiche che hanno valori nulli
    rows_with_null = df[df['Sex'].isnull() | df['Health'].isnull()]
    print("\nRighe con valori Null:")
    print(rows_with_null)
else:
    print("Nessun valore Null trovato nelle colonne Sex e Health.")

# ---------------------------------------------------------
# 3. Controllo valori "Spayed" e "Neutered"
# ---------------------------------------------------------
print("\n2. CONTROLLO VALORI 'Spayed' / 'Neutered'")

bad_values = ['Spayed', 'Neutered']

# Filtra le righe dove Sex O Health contengono quei valori
# Usiamo .isin() per cercare corrispondenze esatte
mask_sex = df['Sex'].isin(bad_values)
mask_health = df['Health'].isin(bad_values)

problematic_rows = df[mask_sex | mask_health]

if not problematic_rows.empty:
    print(f"Attenzione: Trovate {len(problematic_rows)} righe con valori 'Spayed' o 'Neutered':")
    print(problematic_rows)
else:
    print("Ottimo! Nessun valore 'Spayed' o 'Neutered' trovato.")

# ---------------------------------------------------------
# 4. (Opzionale) Visualizzazione rapida dei valori unici
# ---------------------------------------------------------
print("\n### Riepilogo Valori Unici Attuali ###")
print(f"Sex: {df['Sex'].unique()}")
print(f"Health: {df['Health'].unique()}")
```

# Verifica ulteriore dei valori per le colonne Sex e Health
Questo snippet di codice effettua i seguenti controlli:
1) Se nella colonna sex dell'animale è presente come valore "Female" e nella colonna Health il valore è pari None, sostituire quel valore con "Stayed".
2) Se nella colonna sex dell'animale è presente come valore "Male" e nella colonna Health il valore è pari None, sostituire quel valore con "Neutered",

```python
print("STAMPA DEL DATASET PRIMA DELLA MODIFICA")
print(df)
print("-" * 30)

# ---------------------------------------------------
# 2. Applicazione della logica
# ---------------------------------------------------

# REGOLA 1: Se Sex è 'Female' E Health è Null -> Sostituisci con 'Spayed'
# Condizione: (Colonna Sex == Female) AND (Colonna Health è nulla)
condition_female = (df['Sex'] == 'Female') & (df['Health'].isnull())
df.loc[condition_female, 'Health'] = 'Spayed'

# REGOLA 2: Se Sex è 'Male' E Health è Null -> Sostituisci con 'Neutered'
# Condizione: (Colonna Sex == Male) AND (Colonna Health è nulla)
condition_male = (df['Sex'] == 'Male') & (df['Health'].isnull())
df.loc[condition_male, 'Health'] = 'Neutered'

# ---------------------------------------------------
# 3. Risultato
# ---------------------------------------------------
print("\nSTAMPA DEL DATASET DOPO LA MODIFICA")
print(df)
```

# Ricontrollo nella colonna Health se sono presenti ancora valori Null/None

```python
print("\n Controllo ulteriore per la colonna 'Health'")

# Filtriamo le righe dove Health è ancora Null
rows_with_nulls = df[df['Health'].isnull()]

if not rows_with_nulls.empty:
    print(f"Attenzione: trovati {len(rows_with_nulls)} valori mancanti.")
    print("Ecco il dettaglio delle colonne interessate:")

    # Selezioniamo SOLO le colonne Sex e Health da visualizzare
    print(rows_with_nulls[['Sex', 'Health']])
else:
    print("Ottimo! Nessun valore Null residuo nella colonna Health.")
```

# Controllo se nelle colonne latitude, longitude e geopoint sono presenti valori Null o 0
Questo controllo viene effettuato poichè sarebbe inutile non possedere le coordinate geografiche di dove è stato trovato o di dove proviene l'animale

```python
cols_to_check = ['latitude', 'longitude', 'geopoint']

print(f"### Controllo anomalie (Null o 0) in: {cols_to_check} ###\n")

check_null = df[cols_to_check].isnull()

check_zero = df[cols_to_check] == 0

mask = check_null | check_zero


problematic_rows = df[mask.any(axis=1)]

if not problematic_rows.empty:
    print(f"Trovate {len(problematic_rows)} righe con valori Null o Zero:")


    print(problematic_rows[cols_to_check])
else:
    print("Nessun valore Null o Zero trovato nelle colonne specificate.")
```

# Rimozione del punto iniziale dai valori delle colonne "outcome_is_dead", "outcome_is_other", "outcome_is_alive"

```python
cols_to_clean = ['outcome_is_dead', 'outcome_is_other', 'outcome_is_alive']

# Rimuove qualsiasi carattere non alfanumerico iniziale
pattern = r'^[^\w\d]+'

for col in cols_to_clean:
    if col in df.columns:
        df[col] = df[col].fillna('').astype(str).str.replace(pattern, '', regex=True).str.strip()

print(df[cols_to_clean])
```

# Sostituzione dei valori NULL con il valore Unknown per la colonna "Reason for Intake" dato che si tratta di una stringa

```python
# Rinomina la colonna per togliere gli spazi
df = df.rename(columns={'Reason for Intake': 'reason_for_Intake'})

col_name = 'reason_for_Intake'  # Assicurati che il nome corrisponda a quello nel tuo DF

if col_name in df.columns:
    # 1. Sostituisce i valori vuoti (NaN) con "Unknown"
    df[col_name] = df[col_name].fillna('Unknown')

    # 2. Sostituisce l'eventuale stringa di testo "NULL" con "Unknown"
    # (Utile se il CSV conteneva la parola scritta invece che una cella vuota)
    df[col_name] = df[col_name].replace('NULL', 'Unknown')

    # 3. Restituisce e stampa il tipo della variabile
    tipo_variabile = df[col_name].dtype
    print(f"Tipo della variabile '{col_name}': {tipo_variabile}")

    # Verifica visiva dei primi valori
    print("Stampa della colonna dopo la modifica: \n", df[col_name])
else:
    print(f"Colonna '{col_name}' non trovata.")

# Ora puoi accedere con il nome che preferivi
print(df['reason_for_Intake'].dtype)
```
